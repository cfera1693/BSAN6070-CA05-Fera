# **kNN ALGORITHM**

```
#Import the necessary packages. 
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```
#Read the dataset into a DataFrame and remove Label column. 
movies=pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')
movies=movies.iloc[:,:10]
movies
```

```
#Divide our dataset into X and y. 
X=movies.iloc[:,2:]
y=movies['Movie Name']
```

```
#Perform the NearestNeighbors model including 5 number of neighbors and fit to X.
neigh=NearestNeighbors(n_neighbors=5).fit(X)
```

```
#For this specific case, make a list with the movie "attributes" of the post.
the_post = [7.2, 1, 1, 0, 0, 0, 0, 1]
```

```
#Extract the indices of the movies that are similar to the post.
distance, index = neigh.kneighbors([the_post])
```

```
#Print the indices. 
print(index)
```

```
#Create a function that recommends movies given a set of movie "attributes".
def recommend_movies(movie_features_vector):
  distance, index = neigh.kneighbors([movie_features_vector])
  j=0
  for i in index[0]:
    j=j+1
    print('The recomendation No.',j,': ', movies['Movie Name'].iloc[i])
```

```
#Use the function and make the recommendation for the post. 
recommend_movies(the_post)
```
